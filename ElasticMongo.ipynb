```python
pubtime = es.search(index = "covid11", body = {
    "query": {
        "bool": {
            "must" : [
                {"match": {"source_x": "PMC" }},
                {"match": {"abstract": "COVID-19 risk factors"}}
            ],
            
            "filter": [
                {"range" : { "publish_time": {"gte": "2020-01-01"}}}
            ]
        }
    }
})
```

```python
from elasticsearch import Elasticsearch
res= es.search(index='metafile',body={
        'query':{
            'match':{
                "title":"COVID-19 risk factors"
            }
        }
    })
for hit in res['hits']['hits']:
    print(hit['_source']['title']) 
    print(hit['_score'])
    print('**********************')
```
